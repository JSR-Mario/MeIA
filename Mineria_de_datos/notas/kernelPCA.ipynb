```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos: Separar medias lunas
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, random_state=123)
plt.scatter(X[y==0, 0], X[y==0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', marker='o', alpha=0.5)
plt.show()

# print(y==0) esto hace basicamente un map de la funcion sobre el array por lo que regresa un array lleno de booleans con la condicion 
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,4))
ax[0].scatter(X_pca[y==0,0],X_pca[y==0,1],color='red',marker='^',alpha=0.5)
ax[0].scatter(X_pca[y==1,0],X_pca[y==1,1],color='blue',marker='o',alpha=0.5)
ax[1].scatter(X_pca[y==0,0],np.zeros((50,1))+.02,color='red',marker='^',alpha=0.5)
ax[1].scatter(X_pca[y==1,0],np.zeros((50,1))-.02,color='blue',marker='o',alpha=0.5)
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_ylim([-1,1])
ax[0].set_xlabel('PC1')
plt.show()
```

```python
# Datos: Separar medias lunas
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, random_state=123)
```

```python
# Con KPCA de Python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
ax[0].scatter(X_kpca[y==0,0],X_kpca[y==0,1],color='red',marker='^',alpha=0.5)
ax[0].scatter(X_kpca[y==1,0],X_kpca[y==1,1],color='blue',marker='o',alpha=0.5)
ax[1].scatter(X_kpca[y==0,0],np.zeros((50,1)),color='red',marker='^',alpha=0.5)
ax[1].scatter(X_kpca[y==1,0],np.zeros((50,1)),color='blue',marker='o',alpha=0.5)
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_ylim([-1,1])
ax[0].set_xlabel('PC1')
plt.show()
```

```python
# Datos: círculos concéntricos
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=3000, random_state=123, noise=0.1, factor=0.2)
plt.scatter(X[y==0, 0], X[y==0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', marker='o', alpha=0.5)
plt.show()
```

```python
# Intentar con PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,4))
ax[0].scatter(X_pca[y==0,0],X_pca[y==0,1],color='red',marker='^',alpha=0.5)
ax[0].scatter(X_pca[y==1,0],X_pca[y==1,1],color='blue',marker='o',alpha=0.5)
ax[1].scatter(X_pca[y==0,0],np.zeros((1500,1))+.02,color='red',marker='^',alpha=0.5)
ax[1].scatter(X_pca[y==1,0],np.zeros((1500,1))-.02,color='blue',marker='o',alpha=0.5)
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_ylim([-1.5,1.5])
ax[0].set_xlabel('PC1')
plt.show()
```

```python
# Con KPCA de Python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components=2,kernel='rbf',gamma=15)
X_kpca = kpca.fit_transform(X)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
ax[0].scatter(X_kpca[y==0,0],X_kpca[y==0,1],color='red',marker='^',alpha=0.5)
ax[0].scatter(X_kpca[y==1,0],X_kpca[y==1,1],color='blue',marker='o',alpha=0.5)
ax[1].scatter(X_kpca[y==0,0],np.zeros((1500,1)),color='red',marker='^',alpha=0.5)
ax[1].scatter(X_kpca[y==1,0],np.zeros((1500,1)),color='blue',marker='o',alpha=0.5)
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_ylim([-2,2])
ax[0].set_xlabel('PC1')
plt.show()
```

```python
import matplotlib.pyplot as plt

for gamma in [1, 5, 10, 15, 20]:
    kpca = KernelPCA(n_components=2, kernel='rbf', gamma=gamma)
    X_kpca = kpca.fit_transform(X)
    plt.figure()
    plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='coolwarm', s=5)
    plt.title(f'KPCA con gamma={gamma}')
    plt.show()
```
